## 1) Descarga de datos

### 1.1) Top escuchadas del artista

```python
# URL del artirusa
artista = 'https://open.spotify.com/artist/3CAhiUHkUYT1mFtVHM9SHA?si=q7wwfc2zTMG-RSOoxkv5aQ'

# Query
top_tracks = spotify.artist_top_tracks(artist_id=artista)
top_tracks
```

Visualización top 10 escuchadas

```python
# Creación de un DF vacío
top_escuchadas  = pd.DataFrame(columns=["Nombre del artista", "Nombre del album", "Fecha de lanzamiento", "Total de tracks",
                                        "Nombre de la cancion", "duracion", "Popularidad", "ID_song", "ID_artista"])
single = top_escuchadas.copy()


for i in range(0,10):
    # Integración de los datos
    single["Nombre del artista"] = [top_tracks["tracks"][i]["artists"][0]["name"]]
    single["Nombre del album"] = [top_tracks["tracks"][i]["album"]["name"]]
    single["Fecha de lanzamiento"] = [top_tracks["tracks"][i]["album"]["release_date"]]
    single["Total de tracks"] = [top_tracks["tracks"][i]["album"]["total_tracks"]]
    single["Nombre de la cancion"] = [top_tracks["tracks"][i]["name"]]
    single["duracion"] = [top_tracks["tracks"][i]["duration_ms"]]
    single["duracion"] = single ["duracion"]/1000
    single["Popularidad"] =  [top_tracks["tracks"][i]["popularity"]]
    single["ID_song"] =[top_tracks["tracks"][i]["id"]]
    single["ID_artista"] = [top_tracks["tracks"][i]["artists"][0]["href"]]


    #Concatenación
    top_escuchadas = pd.concat([top_escuchadas, single], axis=0, ignore_index=True) #Recordemos axis = 0 es filas

top_escuchadas
```

### 1.3) Artistas similares

```python
# Creación de los DFs
artistas_similares = pd.DataFrame(columns = ["Nombre", "Followers", "Popularidad", "Género 1", "Género 2", "Género 3"])
similar= artistas_similares.copy()

similares =  spotify.artist_related_artists(artist_id= artista) # Por default te da los 20 más vinculados. Podriamos pedir los 20 más vinculados de estos 20 para tenner los 400 (suponiendo que no se reptien)

for i in range(0,20):
    similar["Nombre"] =[similares["artists"][i] ["name"]]
    similar["Followers"] =[similares["artists"][i] ["followers"]["total"]]
    similar["Popularidad"] = [similares["artists"][i] ["popularity"]]
    similar["Género 1"] = [similares["artists"][i] ["genres"][0]]
    similar["Género 2"] = [similares["artists"][i] ["genres"][1]]
    similar["Género 3"] = [similares["artists"][i] ["genres"][2]]

    #concaneacion
    artistas_similares = pd.concat([artistas_similares, similar], axis=0, ignore_index=True)

artistas_similares
```
